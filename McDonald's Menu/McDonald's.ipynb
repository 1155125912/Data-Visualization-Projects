```python
import pandas as pd
import plotly.offline as po
po.init_notebook_mode(connected=True)
import plotly.plotly as py
import plotly.graph_objs as go
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading file
df = pd.read_csv("/Users/vaibhaviawghate/Documents/Data Projects/McDonald's Menu/menu.csv")
df.head()
```

```python
f, axes = plt.subplots(2, 2, figsize=(25,12))
sns.factorplot(x="Category", y="Cholesterol (% Daily Value)", data=df, kind="swarm", ax = axes[0,0])
sns.factorplot(x="Category", y="Carbohydrates", data=df, kind="swarm", ax = axes[0,1])
sns.factorplot(x="Category", y="Protein", data=df, kind="swarm", ax = axes[1,0])
sns.factorplot(x="Category", y="Total Fat", data=df, kind="swarm", ax = axes[1,1])
plt.setp(axes, yticks=[])
f.tight_layout()
plt.close(5)
plt.close(4)
plt.close(3)
plt.close(2)
```

```python
trace = go.Scatter(
    y = df['Cholesterol (% Daily Value)'].values,
    x = df['Item'].values,
    mode='markers',
    marker=dict(
        size= df['Cholesterol (% Daily Value)'].values,
        color = df['Cholesterol (% Daily Value)'].values,
        colorscale='Portland',
        showscale=True
    ),
)
data = [trace]

layout= go.Layout(
    autosize= True,
    title= 'Cholesterol (% Daily Value) per Item on the Menu',
    xaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False
    ),
    yaxis=dict(
        title= 'Cholesterol (% Daily Value)',
        ticklen= 5,
        gridwidth= 2,
        showgrid=False,
        zeroline=False,
        showline=False
    ),
    showlegend= False,
    hovermode = 'closest'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig,filename='scatterChol')
```
